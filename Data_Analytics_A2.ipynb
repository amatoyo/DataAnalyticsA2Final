```python
# Regression Analysis

# Predict App Sessions
X_sessions = df.drop(columns=['User ID', 'App Sessions', 'Distance Travelled (km)', 'Calories Burned'])
y_sessions = df['App Sessions']

# Train the regression model
reg_model_sessions = LinearRegression()
reg_model_sessions.fit(X_sessions, y_sessions)

# Make predictions
y_pred_sessions = reg_model_sessions.predict(X_sessions)

# Evaluate the model
mse_sessions = mean_squared_error(y_sessions, y_pred_sessions)
r2_sessions = r2_score(y_sessions, y_pred_sessions)

print("App Sessions - Mean Squared Error:", mse_sessions)
print("App Sessions - R^2 Score:", r2_sessions)

# Predict Distance Travelled
X_distance = df.drop(columns=['User ID', 'App Sessions', 'Distance Travelled (km)', 'Calories Burned'])
y_distance = df['Distance Travelled (km)']

# Train the regression model
reg_model_distance = LinearRegression()
reg_model_distance.fit(X_distance, y_distance)

# Make predictions
y_pred_distance = reg_model_distance.predict(X_distance)

# Evaluate the model
mse_distance = mean_squared_error(y_distance, y_pred_distance)
r2_distance = r2_score(y_distance, y_pred_distance)

print("\nDistance Travelled - Mean Squared Error:", mse_distance)
print("Distance Travelled - R^2 Score:", r2_distance)

# Predict Calories Burned
X_calories = df.drop(columns=['User ID', 'App Sessions', 'Distance Travelled (km)', 'Calories Burned'])
y_calories = df['Calories Burned']

# Train the regression model
reg_model_calories = LinearRegression()
reg_model_calories.fit(X_calories, y_calories)

# Make predictions
y_pred_calories = reg_model_calories.predict(X_calories)

# Evaluate the model
mse_calories = mean_squared_error(y_calories, y_pred_calories)
r2_calories = r2_score(y_calories, y_pred_calories)

print("\nCalories Burned - Mean Squared Error:", mse_calories)
print("Calories Burned - R^2 Score:", r2_calories)

```

```python
# Clustering

# Standardize the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df.drop(columns=['User ID']))

# Apply PCA to reduce dimensionality for visualization
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(df_scaled)

# Add the cluster labels to the original DataFrame
df['Cluster'] = clusters

# Plot the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_pca[:, 0], y=df_pca[:, 1], hue=clusters, palette='viridis')
plt.title('User Clusters')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()

# Analyzing Clusters
cluster_summary = df.groupby('Cluster').mean()
print("\nCluster Summary:\n", cluster_summary)
```
